B) Remove the units from some of the attributes and only keep the numerical values (for

```python
values_missed = data_set.isnull().sum()
print(values_missed)
```

I have replaced missing values with the mean

```python
#mileage
data_set['Mileage'] = data_set['Mileage'].str.replace(r'\D', '', regex=True)
data_set['Mileage'] = pd.to_numeric(data_set['Mileage'], errors='coerce')
data_set['Mileage'].fillna(data_set['Mileage'].mean(), inplace=True)

#Engine
data_set['Engine'] = data_set['Engine'].str.replace(r'\D', '', regex=True)
data_set['Engine'] = pd.to_numeric(data_set['Engine'], errors='coerce')
data_set['Engine'].fillna(data_set['Engine'].mean(), inplace=True)

#power
data_set['Power'] = data_set['Power'].str.replace(r'\D', '', regex=True)
data_set['Power'] = pd.to_numeric(data_set['Power'], errors='coerce')
data_set['Power'].fillna(data_set['Power'].mean(), inplace=True)

#seats
data_set['Seats'].fillna(data_set['Seats'].mean(), inplace=True)
data_set.drop(columns=['New_Price'], inplace=True)

updated_values = data_set.isnull().sum()

print(updated_values)
```

The key features of a used car are represented by the columns "Mileage," "Power," "Seats," and "Engine." These attributes are typically vital when evaluating a vehicle’s price, condition, and overall performance. Removing these columns would reduce the dimensionality of the dataset which can lead to inaccuarte predivtions. However, keeping them can improve the clarity and accuray of the analysis. Although missing values can be filled using imputation techniques, the data might still be outdated or inaccurate.
I have replaced the missing values with mean to retain the column Power, Engine, Seats, and Mileage.
